```python
import sys
import matplotlib.pyplot as plt
```

```python
logpath = "iterative_refinement/experiments/sr_ffhq_211028_123130/logs/train.log"
# logpath = "iterative_refinement/experiments/sr_ffhq_211023_120750/logs/train.log"

train_losses = []
val_losses = []
with open(logpath, 'r') as f:
    contents = f.read()
    for chunk in contents.split("l_pix: ")[1:]:
        train_losses.append(float(chunk[:10]))
    for chunk in contents.split("PSNR: ")[1:]:
        val_losses.append(float(chunk[:10]))
        
plt.plot(train_losses)
plt.show()
plt.plot(val_losses)
plt.show()
```
